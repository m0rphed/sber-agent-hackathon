```python
import asyncio
import inspect
import json
from pathlib import Path
from typing import Any

from langgraph_app.api.yazzh_final_wip import ApiClientUnified


# ============================================================
# Ğ”Ğ¸Ñ€ĞµĞºÑ‚Ğ¾Ñ€Ğ¸Ñ Ğ¸ Ñ…ĞµĞ»Ğ¿ĞµÑ€Ñ‹ Ğ´Ğ»Ñ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ JSON
# ============================================================

JSON_DIR = Path('yazzh_json')
JSON_DIR.mkdir(exist_ok=True)


def _get_next_index(func_name: str) -> int:
    """
    ĞĞ°Ğ¹Ñ‚Ğ¸ ÑĞ»ĞµĞ´ÑƒÑÑ‰Ğ¸Ğ¹ ÑĞ²Ğ¾Ğ±Ğ¾Ğ´Ğ½Ñ‹Ğ¹ Ğ¸Ğ½Ğ´ĞµĞºÑ Ğ´Ğ»Ñ Ñ„Ğ°Ğ¹Ğ»Ğ° Ğ˜Ğ¼ÑĞ¤ÑƒĞ½ĞºÑ†Ğ¸Ğ¸_N.json
    """
    pattern = f'{func_name}_*.json'
    existing = list(JSON_DIR.glob(pattern))
    if not existing:
        return 1

    max_index = 0
    for p in existing:
        stem = p.stem  # Ğ½Ğ°Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€, get_news_3
        parts = stem.split('_')
        if len(parts) >= 2 and parts[-1].isdigit():
            n = int(parts[-1])
            if n > max_index:
                max_index = n

    return max_index + 1


def _get_function_source(func_name: str) -> str:
    """
    Ğ’Ğ·ÑÑ‚ÑŒ Ğ¸ÑÑ…Ğ¾Ğ´Ğ½Ñ‹Ğ¹ Ñ‚ĞµĞºÑÑ‚ Ğ¼ĞµÑ‚Ğ¾Ğ´Ğ° ApiClientUnified.<func_name>.
    Ğ•ÑĞ»Ğ¸ Ğ¼ĞµÑ‚Ğ¾Ğ´ Ğ½Ğµ Ğ½Ğ°Ğ¹Ğ´ĞµĞ½ Ğ¸Ğ»Ğ¸ Ğ¸ÑÑ…Ğ¾Ğ´Ğ½Ğ¸Ğº Ğ½ĞµĞ´Ğ¾ÑÑ‚ÑƒĞ¿ĞµĞ½ â€” Ğ²ĞµÑ€Ğ½ÑƒÑ‚ÑŒ Ğ¿ÑƒÑÑ‚ÑƒÑ ÑÑ‚Ñ€Ğ¾ĞºÑƒ.
    """
    func_obj = getattr(ApiClientUnified, func_name, None)
    if func_obj is None:
        return ''

    try:
        return inspect.getsource(func_obj)
    except OSError:
        return ''


def _normalize_result(result: Any) -> Any:
    """
    ĞŸÑ€Ğ¸Ğ²ĞµÑÑ‚Ğ¸ Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚ Ğº JSON-ÑĞµÑ€Ğ¸Ğ°Ğ»Ğ¸Ğ·ÑƒĞµĞ¼Ğ¾Ğ¼Ñƒ Ğ²Ğ¸Ğ´Ñƒ.
    Ğ•ÑĞ»Ğ¸ Ğ² Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ğµ ĞµÑÑ‚ÑŒ ĞºĞ»ÑÑ‡ 'json' Ğ¸Ğ»Ğ¸ 'data', Ğ¾Ñ‚Ğ´Ğ°Ñ‚ÑŒ Ğ¸Ñ….
    """
    if isinstance(result, dict):
        if 'json' in result:
            return result['json']
        if 'data' in result:
            return result['data']
    return result


def save_result_json(func_name: str, result: Any) -> Path:
    """
    Ğ¡Ğ¾Ñ…Ñ€Ğ°Ğ½Ğ¸Ñ‚ÑŒ Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚ Ğ²Ñ‹Ğ·Ğ¾Ğ²Ğ° API Ğ² JSON-Ñ„Ğ°Ğ¹Ğ» Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚Ğ°:

    {
      "function_name": "...",
      "function_source": "...",
      "result": {...}
    }

    Ğ˜Ğ¼Ñ Ñ„Ğ°Ğ¹Ğ»Ğ°: <func_name>_N.json, Ğ³Ğ´Ğµ N â€” ÑÑ‡Ñ‘Ñ‚Ñ‡Ğ¸Ğº.
    """
    idx = _get_next_index(func_name)
    file_path = JSON_DIR / f'{func_name}_{idx}.json'

    payload = {
        'function_name': func_name,
        'function_source': _get_function_source(func_name),
        'result': _normalize_result(result),
    }

    with file_path.open('w', encoding='utf-8') as f:
        json.dump(payload, f, ensure_ascii=False, indent=2, default=str)

    return file_path


# ============================================================
# ĞĞ±Ñ‘Ñ€Ñ‚ĞºĞ° Ğ½Ğ°Ğ´ ApiClientUnified._get_request
# Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ Ğ²Ğ¾Ğ¾Ğ±Ñ‰Ğµ Ğ½Ğµ Ñ‚Ñ€Ğ¾Ğ³Ğ°Ñ‚ÑŒ Ñ‚ĞµÑÑ‚Ñ‹
# ============================================================

_original_get_request = ApiClientUnified._get_request


async def _wrapped_get_request(
    self,
    func_name: str,
    url: str,
    params: dict[str, Any] | None = None,
) -> Any:
    """
    ĞĞ±Ñ‘Ñ€Ñ‚ĞºĞ° Ğ½Ğ°Ğ´ Ğ¸ÑÑ…Ğ¾Ğ´Ğ½Ñ‹Ğ¼ _get_request:
    - Ğ²Ñ‹Ğ·Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ¾Ñ€Ğ¸Ğ³Ğ¸Ğ½Ğ°Ğ»ÑŒĞ½Ñ‹Ğ¹ Ğ¼ĞµÑ‚Ğ¾Ğ´
    - ÑĞ¾Ñ…Ñ€Ğ°Ğ½ÑĞµÑ‚ JSON Ğ² Ñ„Ğ°Ğ¹Ğ» Ğ˜Ğ¼ÑĞ¤ÑƒĞ½ĞºÑ†Ğ¸Ğ¸_N.json
    """
    result = await _original_get_request(self, func_name, url, params)
    try:
        save_result_json(func_name, result)
    except Exception as e:
        # Ğ›Ğ¾Ğ³ Ğ¾ÑˆĞ¸Ğ±Ğ¾Ğº ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ñ, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ Ğ½Ğµ Ğ»Ğ¾Ğ¼Ğ°Ñ‚ÑŒ ÑĞ°Ğ¼Ğ¸ Ğ·Ğ°Ğ¿Ñ€Ğ¾ÑÑ‹
        print(f'[save_result_json ERROR] {func_name}: {e}')
    return result


# ĞŸĞ¾Ğ´Ğ¼ĞµĞ½ÑĞµĞ¼ Ğ¼ĞµÑ‚Ğ¾Ğ´ Ğ² ĞºĞ»Ğ°ÑÑĞµ
ApiClientUnified._get_request = _wrapped_get_request  # type: ignore[attr-defined]


# ============================================================
# Ğ—Ğ°Ğ¿ÑƒÑĞº Ğ²ÑĞµÑ… Ñ‚ĞµÑÑ‚Ğ¾Ğ² Ğ² Ñ†Ğ¸ĞºĞ»Ğµ
# ============================================================

async def run_all_tests() -> None:
    """
    ĞŸĞ¾ÑĞ»ĞµĞ´Ğ¾Ğ²Ğ°Ñ‚ĞµĞ»ÑŒĞ½Ğ¾ Ğ¿Ñ€Ğ¾Ğ³Ğ½Ğ°Ñ‚ÑŒ Ğ²ÑĞµ Ñ‚ĞµÑÑ‚Ñ‹ Ğ¸Ğ· ÑĞ¿Ğ¸ÑĞºĞ° tests.
    ĞšĞ°Ğ¶Ğ´Ñ‹Ğ¹ Ğ²Ñ‹Ğ·Ğ¾Ğ² client.<method>() Ğ²Ğ½ÑƒÑ‚Ñ€Ğ¸ Ñ‚ĞµÑÑ‚Ğ°
    Ğ°Ğ²Ñ‚Ğ¾Ğ¼Ğ°Ñ‚Ğ¸Ñ‡ĞµÑĞºĞ¸ ÑĞ¾Ñ…Ñ€Ğ°Ğ½Ğ¸Ñ‚ JSON Ñ‡ĞµÑ€ĞµĞ· Ğ¾Ğ±Ñ‘Ñ€Ğ½ÑƒÑ‚Ñ‹Ğ¹ _get_request.
    """
    total = len(tests)
    for idx, test_func in enumerate(tests, start=1):
        print(f'[{idx}/{total}] {test_func.__name__}')
        try:
            await test_func()
        except Exception as e:
            print(f'  ERROR in {test_func.__name__}: {e}')

```
